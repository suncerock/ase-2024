```python
import soundfile as sf
import numpy as np
import matplotlib.pyplot as plt
```

```python
y, sr = sf.read("sweep.wav")
y_out, sr = sf.read("output.wav")
y.shape
```

```python
Width = 0.1
Modfreq = 2.0
```

```python
output = np.zeros_like(y)

ya_alt = 0
Delay=Width
DELAY=round(Delay*sr)
WIDTH=round(Width*sr)

MODFREQ=Modfreq/sr
LEN=len(y)
L=2+DELAY+WIDTH*2



for channel in range(y.shape[1]):
    Delayline = np.zeros(L)
    for n in range(y.shape[0]):
        M=MODFREQ
        MOD=np.sin(M*2*np.pi*(n-1))
        TAP=1+DELAY+WIDTH*MOD
        i=int(np.floor(TAP))
        frac=TAP-i
        Delayline=np.hstack([y[n], Delayline[:-1]])

        output[n][channel] = Delayline[i+1]*frac + Delayline[i] - (1 - frac)

output.shape
```

```python
plt.plot(np.arange(len(output)) / sr, y_out[:, 0] - output[:, 0])
plt.plot(np.arange(len(output)) / sr, y_out[:, 1] - output[:, 1])
plt.title("Difference")

plt.xlabel("Time")
plt.ylabel("Difference")
```
